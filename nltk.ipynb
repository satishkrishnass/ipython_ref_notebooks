```python
text = ['hi' ,'this', 'is', 'atihs'] 
def un_w(text):
    twords = set(w.lower() for w in text if w.isalpha())
    wlist  = set(w.lower() for w in nltk.corpus.words.words())
    return sorted(twords.difference(wlist))

un_w(text)

```

```python
entries = nltk.corpus.cmudict.entries()
[entry for entry in entries]
```

```python
## A pronouncing dictionary
for word ,pron in entries:
    if len(pron) ==3:
        p1,p2,p3 = pron
        if p1=='P' and p3 == 'T':
            print(word,p2)
```

```python
entries = nltk.corpus.cmudict.entries()
for word, pron in entries:
    if word.lower().__contains__('nicks'):
        print(word,pron)
```

```python
syllable = ['N', 'IH2', 'K', 'S']
[word for word,pron in entries if pron[-4:] == syllable]
```

```python
[word[:2] for word,pron in entries if pron[0]=='K' and word[0] != 'k']
```

```python
def stress(pron):
    return [char for phone in pron for char in phone if char.isdigit()]

[ w for w,pron in entries if stress(pron)== ['0', '1', '0', '2', '0']]
```

```python
p3 = [(pron[0]+'-'+pron[2],word)
     for word,pron in entries
     if pron[0]=='P' and len(pron)==3]

cfd = nltk.ConditionalFreqDist(p3)

for template in cfd.conditions():
    if len(cfd[template])>10:
        words = cfd[template].keys()
        wordlist = ' '.join(words)
        print(template,wordlist[:77])
```

```python
## prononciation dictionary:

prondict = nltk.corpus.cmudict.dict()
text = ['natural','language','processing']

def text_to_speech(text):
    for word in text:
        print('[ ---')
        for phone in prondict[word][0]:
            print(phone)
        print('---]')
            
text_to_speech(text)
```
