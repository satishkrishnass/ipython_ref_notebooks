### Concatenate Data Frames

```python
import pandas as pd
india_weather = pd.DataFrame({
    "city": ['mumbai','delhi','bangalore'],
    "temperature":[32,45,30],
    "humidity" : [80,60,78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city":['new york','chicago','orlando'],
    "temperature":[21,14,35],
    "humidity":[68,65,75]
})

us_weather
```

```python
# concate two dataframes
df = pd.concat([india_weather,us_weather])
df
```

```python
# having proper index
df = pd.concat([india_weather,us_weather], ignore_index=True)
df
```

```python
df = pd.concat([india_weather,us_weather],axis=1)
df
```

### Merge DataFrames

```python
temperature_df = pd.DataFrame({
    "city": ['mumbai','delhi','hyd','bang'],
    "temperature":[32,45,30,40]
})
temperature_df
```

```python
humidity_df=pd.DataFrame({
    'city':['delhi','mumbai','bang'],
    'humidity':[66,44,22]
})

humidity_df
```

```python
# merge two dataframes with out explicitly mention index
df = pd.merge(temperature_df,humidity_df,on='city')
df
```

```python
# Outer-join
df = pd.merge(temperature_df,humidity_df,on='city',how='outer')
df
```

## Pandas Series :



* Pandas Series is a one-dimensional labeled array capable of holding data of any type (integer, string, float, python objects, etc.). The axis labels are collectively called index. Pandas Series is nothing but a column in an excel sheet

```python
# How to create a series from a list, numpy array and dict?


import numpy as nm
myarr = nm.arange(26)
mylist = list('abcedfghijklmnopqrstuvwxyz')
z = zip(myarr,mylist)
mydict = dict(z)

# solution
ser1 = pd.Series(mylist)
ser2 = pd.Series(myarr)
ser3 = pd.Series(mydict)

print(ser1.head())
```

```python
# How to convert the index of a series into a column of a dataframe?

mylist = list('abcedfghijklmnopqrstuvwxyz')
ser = pd.Series(mylist)
print(ser.head())
df = ser.to_frame()  # to_frame is used to convert series to dataframe
df = df.reset_index() 

# Pandas reset_index() is a method to reset index of a Data Frame.
# reset_index() method sets a list of integer ranging from 0 to length of data
# as index.
print(df.head())
```

## Python zip():



* The zip() function takes iterables and aggregates their elements based on their order in the iterable.

* input  : iterables: Any iterables such as lists, strings, etc.
* output : An iterator of tuples based on the input iterables.

* Note: An Iterable is an object which can be repeated over and over. It generates an iterator when passed to iter() method.

*  iter() -- https://www.programiz.com/python-programming/methods/built-in/iter



```python
# example:


names = ['satish','chetan','suda']
marks = [100,98,99]

zipped_iterator = zip(names,marks)
print(type(zipped_iterator))
for i in zipped_iterator:
    print(type(i))
    break
    
#The zip() function returns an iterator. It is important to 
#convert that iterator into an iterable such as a list, dict, set, etc.

# converting iterator to dict
zipped_values = dict(zipped_iterator)

```

```python
import numpy as nm
myarr = nm.arange(26)
mylist = list('abcedfghijklmnopqrstuvwxyz')

mydict = dict(zip(myarr,mylist))
print(mydict)
```

### Pandas DataFrame:

* Pandas DataFrame is two-dimensional size-mutable, potentially heterogeneous tabular data structure with labeled axes (rows and columns). A Data frame is a two-dimensional data structure, i.e., data is aligned in a tabular fashion in rows and columns. Pandas DataFrame consists of three principal components, the data, rows, and columns.

####  Iterating over rows 
* In order to iterate over rows, we can use three function iteritems(), iterrows(), itertuples() . These three function will help in iteration over rows.

```python
# importing pandas as pd
import pandas as pd

# dictionary of lists
dict = {'name':["aparna", "pankaj", "sudhir", "Geeku"],
        'degree': ["MBA", "BCA", "M.Tech", "MBA"],
        'score':[90, 40, 80, 98]}
 
# creating a dataframe from a dictionary 
df = pd.DataFrame(dict)
print(df)
print()
# iterating over rows using iterrows() function 
for i,j in df.iterrows():
    print(i,j)
    print()
```
