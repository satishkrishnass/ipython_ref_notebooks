```python
import numpy as np
import pylab
import scipy.stats as stats
import statistics as st

std = np.random.normal(loc=20,scale=5,size =100)
stats.probplot(std,dist='norm',plot = pylab)
pylab.show()

uni= np.random.uniform(low=-20,high=20,size = 1000)
stats.probplot(uni,dist='uniform',plot=pylab)
pylab.show()

```
