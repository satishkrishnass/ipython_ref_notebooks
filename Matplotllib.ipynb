### Pyplot tutorial:

*  matplotlib.pyplot is a collection of command style functions that make matplotlib work like MATLAB

* Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels etc.


```python
import matplotlib.pyplot  as plt

plt.plot([1,2,4,6,2])
plt.xlabel('x -label')
plt.ylabel('y-label')
plt.title('title')
plt.grid()
plt.show()
```

* if you provide a singel list or array to the plot() command, matplotlib assumes it is a sequence of y values, and automatically generates the x values for you. Since python ranges start with 0, the default x vector has the same length as y but starts with 0. Hence the x data are [0,1,2,3]

```python
plt.plot([1,2,3,4,-2],[2,4,9,16,4])
plt.ylabel('y-label')
plt.xlabel('x -label')
plt.title('title')
plt.grid()
plt.show()
```

* for every x,y pair of arguments, there is an optional third argument which is the format string that indicates the color and line type of the plot

```python
plt.plot([1,2,3,4,-2],[2,4,9,16,4],'ro')
plt.grid()
plt.show
```

* matplotlib will use numpy arrays also to do numeric procecssing , it is not limited to lists only ( if it is, then it is useless). In fact, all sequences all converted to numpy arrays internally.


```python
import numpy as np
t = np.arange(0.,5.,0.2)

# blue lines, red squares,green triangles, and legend is used to represent  
# all line labels
plt.plot(t,t**2,'b--',label='^2')
plt.plot(t,t**3,'rs',label='^3')
plt.plot(t,t,'g^',label='^1')
plt.grid()
plt.legend()
plt.show()
```

###  Controlling line properties

#### use keyword args

```python
x = [1,2,3,4]
y = [1,4,9,16]
plt.plot(x,y,linewidth=5.9)
plt.show()
```

### use the setp()

```python
x1=[1,2,3,4]
y1=[1,4,9,16]
x2=[1,2,3,4]
y2=[2,4,6,8]
lines = plt.plot(x1,y1,x2,y2)
#lines contains 2 lines 
# use keyword args
plt.setp(lines[0],color='r',linewidth=2.0)
plt.setp(lines[1],'color','g','linewidth',3.0)
plt.grid()

```

### working with multiple figures and axes

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)
plt.figure(1)
# the subplot command specifies numrows,numcols
#figure where fignum ranges from 1 to numrows*numcols
plt.subplot(211)
plt.grid()
plt.plot(t1,f(t1),'b--')

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
plt.show()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot([1,2,3])
plt.subplot(212)
plt.plot([4,5,6])
plt.figure(2)
plt.plot([3,6,9])  # creates a subplot(111) by default
plt.figure(1)
plt.subplot(212)
plt.title('just a')
plt.show()
```
